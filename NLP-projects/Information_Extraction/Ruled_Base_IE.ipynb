```python
import re  
import string
import nltk
import spacy
import pandas as pd
import numpy as np 
import  math


from spacy.matcher import Matcher 
from spacy.tokens import Span 
from spacy import displacy 
import en_core_web_sm

pd.set_option('display.max_colwidth', 100)

```

```python


# loading Hearst pattern 
nlp = spacy.load("en_core_web_sm")

text_2 = "Sebastian Thrun is working at Google since 2007. He is a former employee of Salesforce and loved working on NLP, NLTK and Spacy."

# spacy obj
doc = nlp(text_2)
displacy.render(doc, style='dep', jupyter=True, options={'distance':130})
```

```python
displacy.render(doc, style='ent', jupyter=True, options={'distance':130})
```

```python
print(doc.ents) 
print([X for X in doc.sents])
```

```python
# pattern matching 

pattern = [ {'POS': 'PROPN',  'OP':"?"},
            {'POS': 'PROPN'},
            {'LOWER': 'loved', 'OP':"?"},
            {'LOWER': 'is', 'OP':"?"},
            {'LOWER': 'working'},
            {'LOWER': 'at', 'OP':"?"},
            {'LOWER': 'on', 'OP':"?"},
            {'POS': 'PROPN'},
        ]

matcher = Matcher(nlp.vocab)

matcher.add("work_on_pattern", None, pattern)

matches = matcher(doc)


print(matches)
print(doc[matches[0][1]:matches[0][2]])
print(doc[matches[1][1]:matches[1][2]])


```
